## B

```python
student_data=[61,93,75,67,84,95]
s_mean = st.mean(student_data)
s_sd=st.stdev(student_data)
n=len(student_data)
print('Sample mean ='+str(s_mean))
print('sample SD ='+str(s_sd))
print('sample size ='+str(n))
```

```python
p_mean=71  #population mean
s_mean=79.16 #sample mean
s_sd=14 #sample std 
n=6  #sample size
dof= n-1 #degree of freedom 
alpha=0.05 #signoificance leval
```

```python
t_statistics = (s_mean - p_mean)/(s_sd/math.sqrt(n))  #n= sample size
print('The t_statistics is ',t_statistics)

neg_t_critical = t.ppf(alpha/2,dof) #dof to calculate t-critical values
print('The neg t-critical value is'+str(neg_t_critical))

pos_t_critical = t.ppf(1-(alpha/2),dof) #dof to calculate t-critical values
print('The pos t-critical value is'+str(pos_t_critical))
```

```python
p_value = t.sf(abs(t_statistics),dof)*2
print(p_value)
```

```python
# establish the decison rule
if neg_t_critical < t_statistics < pos_t_critical and p_value > alpha:
    print("accept null HO ie no significant difference")
else :
    print("reject null HO ; ie there is significant difference")
```

```python
x=np.arange(-3,3,0.01)
tvals2=t.pdf(x,2)
tvals4=t.pdf(x,4)
tvals8=t.pdf(x,8)
tvals30=t.pdf(x,30)
plt.figure(figsize=(15,5))
plt.xlabel('X')
plt.ylabel('prob. Values')
plt.title('T-Dist Curve')
plt.plot(x,tvals2,x,tvals4,x,tvals8,x,tvals30,marker='*')
plt.grid()
plt.legend(['df=2','df=4','df=8','df=30'],shadow = True,loc='best')
plt.show()
```

## Task 2

```python
sample_machine1=[4.05,4.01,4.04,4.02]
sample_machine2=[4.02,3.99,4.01,3.99,4.00,4.00]
```

```python
x1=st.mean(sample_machine1)
x2=st.mean(sample_machine2)
v1=st.variance(sample_machine1)
v2=st.variance(sample_machine2)
n1=len(sample_machine1)
n2=len(sample_machine2)
df=n1+n2-2    #Degree of freeedom
print('sample mean of machine1 = ',x1)
print('sample mean of machine2 = ',x2)
print('sample variance of machine1 = ',v1)
print('sample variance of machine2 = ',v2)
print('sample size of machine1 = ',n1)
print('sample vsize of machine2 = ',n2)
print('Dgree of freedom = ',df)
```

```python
n=abs(x1-x2)
d=math.sqrt(v1*(n1-1)+v2*(n2-1))*math.sqrt(1/n1+1/n2)
t_statistics=n/d
print('t_statistics',t_statistics)
neg_t_critical = t.ppf((alpha/2),df)
print("The neg t-critical value is ",neg_t_critical )

pos_t_critical = t.ppf(1-(alpha/2),df)
print("The neg t-critical value is ",pos_t_critical )
```

```python
p_value = t.sf(abs(t_statistics),df)*2
print(p_value)
```

```python
if neg_t_critical < t_statistics < pos_t_critical and p_value > alpha:
    print("accept null HO ie no significant difference")
else :
    print("reject null HO ; ie there is significant difference")
```

## SSC

```python
SSC= n1*(x1-x)**2 + n2*(x2-x)**2 + n3*(x3-x)**2 + n4*(x4-x)**2
print('SSC = ',SSC)
```

## SST
