## 3.1.4
stratified sampling with respect to genre

```python
group_Game_df=Game_df.groupby(['Genre'])
genre=Game_df.Genre.unique()
stratified_sample_df=pd.DataFrame()

for i in genre:
    c=group_Game_df.get_group(i)
    sample_data=c.sample(n=3,replace=True,random_state=3)
    stratified_sample_df=stratified_sample_df.append(sample_data,ignore_index=True)
    
stratified_sample_df
```

## 3.1.5
clustered_sampling with respect to 2 genre

```python
group_Game_df=Game_df.groupby(['Genre'])
genre=['Platform','Racing']
clustered_sample_df=pd.DataFrame()

for i in genre:
    c=group_Game_df.get_group(i)
    sample_data=c.sample(n=20,replace=True,random_state=3)
    clustered_sample_df=clustered_sample_df.append(sample_data,ignore_index=True)
    
clustered_sample_df
```
