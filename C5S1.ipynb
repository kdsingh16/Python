```python
import pandas as pd 
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sb
```

```python
car=pd.read_csv("DS1_C5_S1_Descriptive_Analysis_Practice_Car_Data.csv")
car.head()
```

```python
car1 = car.iloc[:,[1,4,11,18,19,72]]
car1.head()
```

```python
car2=car1.dropna()
car2.head(2)
```

## Task 1

```python
#maruti 
m=car2[(car2.Make=='Maruti Suzuki')]
m=m.drop('Make',axis=1)
m.head(2)
```

```python
mean=[]
median=[]
std=[]
cv=[]
for col in m.head():
    col_mean=m[col].mean()
    mean.append(col_mean)
    col_median=m[col].median()
    median.append(col_median)
    col_std=m[col].std()
    std.append(col_std)
    cv.append(col_std/col_mean*100)
    
row_head=['Mean','Median','standard deviation','Cofefficient of variance']
col_names=list(m.columns)
central=pd.DataFrame([mean,median,std,cv],columns=col_names,index=[1,2,3,4])
central.insert(0,'Measures',row_head)
central.round(2)
```
