## TASK - 1

```python
#A
Ho=33.88 #null hypothesis 
Ha =33.88 # Sigificant Hypothesis

x=32.18    #sample mean
mu=33.88   #population mean
s=1.28     #population standard deviation
n=49
alpha=0.05

#c
z_statistic=(x-mu)/(s/math.sqrt(n))
print("The z_statistic is ",z_statistic)

#E
p_value = norm.sf(abs(z_statistic))*2  # two tailed test
print("The p_value is "+str(p_value))

#B
pos_z_critical = norm.ppf(1-alpha/2)  #two tailed test
print("The z-critical value is "+str(pos_z_critical))

neg_z_critical = norm.ppf(alpha/2) # two tailed test
print("The -z-critical value is  "+str(neg_z_critical))
```

## TASK 2

```python
#A
Ho<=85 #null hypothesis 
Ha >85 # Sigificant Hypothesis

x2=87.5    #sample mean
mu2=85   #population mean
s2=2.1     #population standard deviation
n2=32
alpha2=0.05

#c
z_statistic2=(x2-mu2)/(s2/math.sqrt(n2))
print("The z_statistic is ",z_statistic2)

#E
p_value2 = norm.sf(abs(z_statistic2))
print("The p_value is "+str(p_value2))

#B
pos_z_critical2 = norm.ppf(1-alpha2)
print("The z-critical value is "+str(pos_z_critical2))

neg_z_critical2 = norm.ppf(alpha2) 
print("The -z-critical value is  "+str(neg_z_critical2))
```

```python
if z_statistic2<pos_z_critical2 and p_value2>alpha2:
    print("accept null HO ie no significant difference")
else :
    print("reject null HO ; ie there is significant difference")
```
